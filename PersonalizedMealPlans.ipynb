```python
# author: Zhizhen L.
import itertools
import random

fruits_options = [
    {
        'name': 'Berry Chia Pudding',
        'calories': 200,
        'protein': 5,
        'fat': 0,
        'fiber': 10,
        'carbs': 20
    },
    {
        'name': 'Baked Apple Slices',
        'calories': 80,
        'protein': 0,
        'fat': 0,
        'fiber': 4,
        'carbs': 20
    },
    {
        'name': 'Citrus Salad',
        'calories': 70,
        'protein': 1,
        'fat': 0,
        'fiber': 3,
        'carbs': 15
    },
    {
        'name': 'Melon Salad',
        'calories': 60,
        'protein': 1,
        'fat': 0,
        'fiber': 2,
        'carbs': 15
    },
    {
        'name': 'Sugar-Free Mixed Berry Compote',
        'calories': 50,
        'protein': 1,
        'fat': 0,
        'fiber': 4,
        'carbs': 10
    }
]

vegetables_options = [
    {
        'name': 'Roasted Veggie Quinoa Bowl',
        'calories': 300,
        'protein': 8,
        'fat': 0,
        'fiber': 8,
        'carbs': 50
    },
    {
        'name': 'Steamed Asparagus with Lemon',
        'calories': 30,
        'protein': 2,
        'fat': 0,
        'fiber': 2,
        'carbs': 5
    },
    {
        'name': 'Zucchini Noodles with Pesto',
        'calories': 250,
        'protein': 5,
        'fat': 0,
        'fiber': 4,
        'carbs': 30
    },
    {
        'name': 'Cucumber and Tomato Salad',
        'calories': 30,
        'protein': 1,
        'fat': 0,
        'fiber': 2,
        'carbs': 5
    },
    {
        'name': 'Garlic Saut√©ed Spinach',
        'calories': 40,
        'protein': 3,
        'fat': 0,
        'fiber': 2,
        'carbs': 6
    }
]

protein_foods_options = [
    {
        'name': 'Grilled Chicken Breast with Herbs',
        'calories': 150,
        'protein': 30,
        'fat': 0,
        'fiber': 0,
        'carbs': 0
    },
    {
        'name': 'Baked Salmon with Dill',
        'calories': 250,
        'protein': 25,
        'fat': 0,
        'fiber': 0,
        'carbs': 0
    },
    {
        'name': 'Lentil and Vegetable Soup',
        'calories': 250,
        'protein': 15,
        'fat': 0,
        'fiber': 10,
        'carbs': 40
    },
    {
        'name': 'Tofu and Vegetable Stir-Fry',
        'calories': 300,
        'protein': 20,
        'fat': 0,
        'fiber': 6,
        'carbs': 20
    },
    {
        'name': 'Turkey and Avocado Lettuce Wraps',
        'calories': 250,
        'protein': 20,
        'fat': 0,
        'fiber': 5,
        'carbs': 10
    }
]

dairy_options = [
    {
        'name': 'Greek Yogurt Parfait',
        'calories': 200,
        'protein': 15,
        'fat': 0,
        'fiber': 3,
        'carbs': 20
    },
    {
        'name': 'Cottage Cheese and Cucumber',
        'calories': 150,
        'protein': 15,
        'fat': 0,
        'fiber': 1,
        'carbs': 10
    },
    {
        'name': 'Cheese and Veggie Plate',
        'calories': 200,
        'protein': 10,
        'fat': 0,
        'fiber': 4,
        'carbs': 20
    },
    {
        'name': 'Dairy-Free Yogurt Alternative',
        'calories': 150,
        'protein': 5,
        'fat': 0,
        'fiber': 4,
        'carbs': 20
    },
    {
        'name': 'Ricotta and Spinach Stuffed Chicken',
        'calories': 300,
        'protein': 35,
        'fat': 0,
        'fiber': 2,
        'carbs': 5
    }
]

grains_options = [
    {
        'name': 'Brown Rice and Black Bean Bowl',
        'calories': 350,
        'protein': 10,
        'fat': 0,
        'fiber': 8,
        'carbs': 70
    },
    {
        'name': 'Quinoa Stuffed Peppers',
        'calories': 350,
        'protein': 15,
        'fat': 0,
        'fiber': 6,
        'carbs': 65
    },
    {
        'name': 'Whole Wheat Pasta Primavera',
        'calories': 350,
        'protein': 12,
        'fat': 0,
        'fiber': 7,
        'carbs': 70
    },
    {
        'name': 'Barley Salad',
        'calories': 250,
        'protein': 5,
        'fat': 0,
        'fiber': 8,
        'carbs': 50
    },
    {
        'name': 'Vegetable Fried Cauliflower Rice',
        'calories': 300,
        'protein': 15,
        'fat': 0,
        'fiber': 6,
        'carbs': 60
    }
]

meal_options = {
    'fruits': fruits_options,
    'vegetables': vegetables_options,
    'protein_foods': protein_foods_options,
    'dairy': dairy_options,
    'grains': grains_options
}


def main():
    prediction_results = input("What is your results from one of our prediction models? (low/medium/high/very high): ")

    if prediction_results == "low":
        print("That is very good news. However, one of the most effective ways to prevent diabetes is to have a healthy lifestyle.")
        answer_two_questions()  # Ask about vegetarian and lactose intolerance
        generate_random_meal()  # Generate a meal based on preferences
    elif prediction_results == "medium":
        print("Not bad! However, one of the most effective ways to prevent diabetes is to have a healthy lifestyle.")
        answer_two_questions()
        generate_random_meal()
    elif prediction_results == "high" or prediction_results == "very high":
        print("Not the best news but that's okay. We are here for you. Maintaining a healthy lifestyle is among the most important in preventing and managing diabetes.")
        answer_two_questions()
        answer_optional_questions()

def answer_optional_questions(vegetarian, lactose_intolerant):
    answer_optional = input("Do you want to answer optional questions for a personalized meal plan? (y/n): ").lower()

    if answer_optional == 'y':
        ideal_calories = input("What is your ideal calories intake? Enter a value or idk to skip: ")
        ideal_protein = input("What is your ideal protein intake in grams? Enter a value or idk to skip: ")
        ideal_fat = input("What is your ideal fat intake in grams? Enter a value or idk to skip: ")
        ideal_carbs = input("What is your ideal carb intake in grams? Enter a value or idk to skip: ")

        # Convert input to floats if not "idk"
        if ideal_calories != "idk":
            ideal_calories = float(ideal_calories)
        if ideal_protein != "idk":
            ideal_protein = float(ideal_protein)
        if ideal_fat != "idk":
            ideal_fat = float(ideal_fat)
        if ideal_carbs != "idk":
            ideal_carbs = float(ideal_carbs)

        meal_plan = generate_random_meal_plan(vegetarian, lactose_intolerant, meal_options)
        compare_meal_plans(ideal_calories, ideal_protein, ideal_fat, ideal_carbs, meal_plan)
        generate_meal_plan(meal_plan, ideal_calories, ideal_protein, ideal_fat, ideal_carbs)
    elif answer_optional == 'n':
        generate_random_meal()
    else:
        print("Invalid input. Please enter 'y' or 'n'.")
        answer_optional_questions(vegetarian, lactose_intolerant)

def answer_two_questions():
    vegetarian = input("Do you identify as vegetarian? (y/n): ").lower() == 'y'
    lactose_intolerant = input("Do you have lactose intolerance? (y/n): ").lower() == 'y'
    if vegetarian or lactose_intolerant:
        answer_optional_questions(vegetarian, lactose_intolerant)
    else:
        generate_random_meal()

def generate_random_meal():
    meal_plan = {
        'fruits': random.choice(fruits_options),
        'vegetables': None,
        'grains': random.choice(grains_options),
        'protein_foods': random.choice(protein_foods_options),
        'dairy': None
    }

    print("\nGenerated Random Meal:")
    print("Fruits:", meal_plan['fruits']['name'])
    print("Vegetables: None (vegetarian)")
    print("Protein Foods:", meal_plan['protein_foods']['name'])
    print("Dairy: None (lactose intolerant)")
    print("Grains:", meal_plan['grains']['name'])


def generate_random_meal_plan(vegetarian, lactose_intolerant, meal_options):
    meal_plan = {
        'fruits': None,
        'vegetables': None,
        'grains': None,
        'protein_foods': None,
        'dairy': None
    }

    # Generate random meal options for each category
    for category, options in meal_options.items():
        # Exclude vegetables for vegetarians and dairy for lactose intolerant individuals
        if (vegetarian and category == 'vegetables') or (lactose_intolerant and category == 'dairy'):
            continue

        # Select a random meal option from the available options
        random_option = random.choice(options)

        # Add the selected option to the meal plan
        meal_plan[category] = random_option

    return meal_plan

def compare_meal_plans(ideal_calories, ideal_protein, ideal_fat, ideal_carbs, meal_plan):

    best_difference = float('inf')  # Initialize with a high value
    best_plan = None

    # Iterate through all possible meal plan combinations
    for combination in itertools.product(*meal_plan.values()):
        total_calories = sum(item['calories'] for item in combination)
        total_protein = sum(item['protein'] for item in combination)
        total_fat = sum(item['fat'] for item in combination)
        total_carbs = sum(item['carbs'] for item in combination)

        # Calculate differences between user's goals and actual values
        cal_difference = abs(total_calories - ideal_calories) if ideal_calories is not None else 0
        protein_difference = abs(total_protein - ideal_protein) if ideal_protein is not None else 0
        fat_difference = abs(total_fat - ideal_fat) if ideal_fat is not None else 0
        carbs_difference = abs(total_carbs - ideal_carbs) if ideal_carbs is not None else 0

        # Calculate a weighted total difference
        total_difference = cal_difference + protein_difference + fat_difference + carbs_difference

        if total_difference < best_difference:
            best_difference = total_difference
            best_plan = combination

    return best_plan

def generate_meal_plan(meal_plan, ideal_calories, ideal_protein, ideal_fat, ideal_carbs):
    # Print the generated meal plan
    print("\nGenerated Meal Plan:")
    print("Fruits:", meal_plan['fruits']['name'])
    print("Vegetables:", meal_plan['vegetables']['name'])
    print("Protein Foods:", meal_plan['protein_foods']['name'])
    print("Dairy:", meal_plan['dairy']['name'])
    print("Grains:", meal_plan['grains']['name'])


if __name__ == "__main__":
    main()


```
