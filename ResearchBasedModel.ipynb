```python
def calculate_diabetes_risk():

    # Create a map and initialize base risk values for each diabetes type,
   diabetes_risk = {
        1: 0,
        2: 0,
        'gestational': 0
    }

    # Another map, define weight values
    weights = {
        'type_1': 2,
        'type_2': 2,
        'gestational': 2,
        'add': 2,
        'double': 4,
        'triple': 6,
        'smoke_alcohol_stress_hp': 0.5,
        'sleep_age_12': 0.5,
        'sleep_age_18': 0.5
    }

    # Gather user input
    biological_sex = input("What is your biological sex? Male/Female: ").lower()

    if biological_sex == "male":
        diabetes_risk['gestational'] = 0
    else:
        is_pregnant = input("Are you pregnant? Yes/Idk/No: ").lower()

        if is_pregnant == "yes":
            gestational_diabetes_pregnancy = input("Did you have gestational diabetes in your last pregnancy? Yes/No: ").lower()
            if gestational_diabetes_pregnancy == "yes":
                diabetes_risk['gestational'] += weights['gestational']
        elif is_pregnant == "no":
            age = int(input("What is your age? "))
            if 4 <= age <= 7 or 10 <= age <= 14:
                diabetes_risk[1] += weights['double']
            elif 35 <= age < 45:
                diabetes_risk[2] += weights['add']
                diabetes_risk['gestational'] += weights['add']
            elif age >= 45:
                diabetes_risk[2] += weights['double']
                diabetes_risk['gestational'] += weights['double']

    if is_pregnant == "yes":
        is_baby_larger = input("Is your baby larger than 4000 gm? Yes/No: ").lower()
        if is_baby_larger == "yes":
            diabetes_risk['gestational'] += weights['add']
            # Allow calculation for gestational
            pass
        elif is_baby_larger == "no":
            diabetes_risk['gestational'] = 0

    age = int(input("What is your age? "))
    if 4 <= age <= 7 or 10 <= age <= 14:
        diabetes_risk[1] += weights['double']
    elif 35 <= age < 45:
        diabetes_risk[2] += weights['add']
        diabetes_risk['gestational'] += weights['add']
    elif age >= 45:
        diabetes_risk[2] += weights['double']
        diabetes_risk['gestational'] += weights['double']

    weight = float(input("Enter your weight in pounds: "))
    height = float(input("Enter your height in inches: "))
    bmi = (weight / (height ** 2)) * 703

    # Define BMI thresholds
    bmi_threshold1 = 25  # Example threshold
    bmi_threshold2 = 30  # Example threshold

    if bmi > bmi_threshold1:
        diabetes_risk[1] += weights['add']
        diabetes_risk[2] += weights['add']
        diabetes_risk['gestational'] += weights['add']
        if bmi > bmi_threshold2:
            diabetes_risk[1] += weights['double']
            diabetes_risk[2] += weights['double']
            diabetes_risk['gestational'] += weights['double']

    exercise_days = input("How often do you exercise? 0-7 days: ")
    if exercise_days == "0-2" or exercise_days == "0-1":
        diabetes_risk[1] += weights['double']
        diabetes_risk[2] += weights['double']
        diabetes_risk['gestational'] += weights['double']

    family_history = input("Do you have a family history of diabetes? Yes/No/Idk: ").lower()
    if family_history == "yes":
        diabetes_risk[1] += weights['triple']
        diabetes_risk[2] += weights['triple']
        diabetes_risk['gestational'] += weights['triple']
    elif family_history == "idk":
        diabetes_risk[1] += weights['add']
        diabetes_risk[2] += weights['add']
        diabetes_risk['gestational'] += weights['add']

    balanced_diet = input("Do you have a well-balanced diet? Yes/No/Idk: ").lower()
    if balanced_diet == "no":
        diabetes_risk[1] += weights['double']
        diabetes_risk[2] += weights['double']
        diabetes_risk['gestational'] += weights['double']

    identify_as_white = input("Do you identify as white? Yes/No: ").lower()
    if identify_as_white == "yes":
        diabetes_risk[1] += weights['add']
    elif identify_as_white == "no":
        diabetes_risk[2] += weights['add']
        diabetes_risk['gestational'] += weights['add']

    smoke = input("Do you smoke? Yes/No: ").lower()
    if smoke == "yes":
        for key in diabetes_risk:
            diabetes_risk[key] += weights['smoke_alcohol_stress_hp']

    alcohol = input("Do you ingest alcohol? Yes/No: ").lower()
    if alcohol == "yes":
        for key in diabetes_risk:
            diabetes_risk[key] += weights['smoke_alcohol_stress_hp']

    stress = input("Are you stressed? Yes/No: ").lower()
    if stress == "yes":
        for key in diabetes_risk:
            diabetes_risk[key] += weights['smoke_alcohol_stress_hp']

    high_blood_pressure = input("Do you have high blood pressure? Yes/No/Idk: ").lower()
    if high_blood_pressure == "yes":
        for key in diabetes_risk:
            diabetes_risk[key] += weights['smoke_alcohol_stress_hp']

    sleep_hours = float(input("How many hours of sleep do you get? 0-24: "))
    if age <= 12 and sleep_hours < 10:
        for key in diabetes_risk:
            diabetes_risk[key] += weights['sleep_age_12']
    elif 12 < age < 18 and sleep_hours < 8:
        for key in diabetes_risk:
            diabetes_risk[key] += weights['sleep_age_18']
    elif age >= 18 and sleep_hours < 7:
        for key in diabetes_risk:
            diabetes_risk[key] += weights['sleep_age_18']

    # Determine risk levels
    risk_levels = {
        'low': 5,
        'medium': 10,
        'high': 15,
        'very high': 20
    }

    # Determine the risk category for each diabetes type
    type1_risk = min(diabetes_risk[1], max(risk_levels.values()))
    type2_risk = min(diabetes_risk[2], max(risk_levels.values()))
    gestational_risk = min(diabetes_risk['gestational'], max(risk_levels.values()))

    # Map risk levels to risk categories
    risk_categories = {
    'low': 'low',
    'medium': 'medium',
    'high': 'high',
    'very high': 'very high'
    }

  # Determine the risk category for each diabetes type
   type1_category = next(category for category, threshold in risk_levels.items() if type1_risk <= threshold)
   type2_category = next(category for category, threshold in risk_levels.items() if type2_risk <= threshold)
   gestational_category = next(category for category, threshold in risk_levels.items() if gestational_risk <= threshold)

  # Print risk evaluations for type 1, type 2, and gestational diabetes
  return("Risk evaluation for Type 1 Diabetes:", risk_categories[type1_category])
  return("Risk evaluation for Type 2 Diabetes:", risk_categories[type2_category])
  return("Risk evaluation for Gestational Diabetes:", risk_categories[gestational_category])
```
